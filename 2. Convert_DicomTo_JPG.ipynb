```python
import pydicom
import zipfile
import os
from PIL import Image
import numpy as np
import glob
# Đọc file DICOM
output_folder = '/content/drive/MyDrive/Data/New_jpg'
Link_jpg=[]
for link in df['Link data']:
  folder_path = link
  if os.path.isdir(folder_path):
      # Tìm tất cả các file DICOM trong thư mục
      dicom_files = glob.glob(os.path.join(folder_path, '*.dcm'))
      if dicom_files:
          print(f"Found {len(dicom_files)} DICOM files:")
          for file in dicom_files:
            ds = pydicom.dcmread(file)
            if 'PixelData' in ds:
              new_image = ds.pixel_array.astype(float)
              scaled_image = (np.maximum(new_image, 0) / new_image.max()) * 255.0
              scaled_image = np.uint8(scaled_image)
              final_image = Image.fromarray(scaled_image)
              base_name = os.path.splitext(os.path.basename(file))[0]
              jpg_path=output_folder+ "/"+ base_name + '.jpg'
              Link_jpg.append({"Link": jpg_path})
              final_image.save(os.path.join(output_folder, base_name + '.jpg'))
            # final_image.save(os.path.join(output_folder, new_file + ' .jpg'))
            # print(f"Converted {file} to {jpg_file}")
# for link in Link_data:
#   sheet.append([link])
# workbook.save('/content/drive/MyDrive/Data/jpg_data.xlsx')
```
