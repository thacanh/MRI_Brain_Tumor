```python
import os
import pydicom as dicom
import pandas as pd

# Đường dẫn tới thư mục chứa file DICOM
dicom_dir = '/content/drive/MyDrive/Data'
output_excel_path = '/content/drive/MyDrive/Data/dicom_data.xlsx'  # Đường dẫn tới tệp Excel đầu ra

# Danh sách để chứa dữ liệu
all_data = []

# Kiểm tra danh sách các file trong thư mục
files = os.listdir(dicom_dir)

# Duyệt qua từng file trong danh sách
for file in files:
    # Chỉ xử lý các file có đuôi .dcm
    if file.endswith('.dcm'):
        dicom_file_path = os.path.join(dicom_dir, file)

        try:
            # Đọc file DICOM
            ds = dicom.dcmread(dicom_file_path)

            # Tạo dictionary để lưu dữ liệu của tệp hiện tại
            dicom_data = {'File Name': file}

            # Duyệt qua các phần tử dữ liệu và trích xuất tên, giá trị
            for elem in ds:
                dicom_data[elem.name] = elem.value

            # Thêm dictionary vào danh sách
            all_data.append(dicom_data)

        except Exception as e:
            print(f"Không thể đọc file {file}: {e}")

# Tạo DataFrame từ danh sách dữ liệu
df = pd.DataFrame(all_data)

# Lưu DataFrame vào tệp Excel
df.to_excel(output_excel_path, index=False)
print(df)
print(f"Dữ liệu đã được lưu vào {output_excel_path}")
```
